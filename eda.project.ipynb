This project performs Exploratory Data Analysis (EDA) on the South India Postal dataset to understand patterns, distributions, and key insights.

## State-wise Distribution of Post Offices

```python
df['State'].value_counts().plot(kind='bar')
plt.title("Post Offices by State")
plt.xlabel("State")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

## Office Type Distribution

```python
df['Office Type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Office Type Distribution")
plt.ylabel("")
plt.show()
```

## Delivery Status Distribution

```python
df['Delivery Status'].value_counts().plot(kind='bar')
plt.title("Delivery Status Distribution")
plt.xlabel("Delivery Status")
plt.ylabel("Count")
plt.show()
```

## Top 10 Districts with Most Post Offices

```python
df['District'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Districts")
plt.xlabel("District")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

## Conclusion

- Telangana has slightly more post offices than Andhra Pradesh.
- All offices are of type BO.
- Most offices provide delivery services.
- Some districts have higher concentration of post offices.
